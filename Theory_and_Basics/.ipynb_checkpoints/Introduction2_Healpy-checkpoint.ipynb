```python
cl=np.loadtxt("cmb_spectrum.txt")
```

```python
def loglikeli(As,ns):
    theta=get_spectrum(As,ns)
    low=30
    up=1200
    sam=0
    for i in range(low,up):
        like=-(2*i+1)/4*((theta[i]-cl[i])**2/cl[i]**2)
        sam=sam+like
    return sam
```

```python
As=1e-9
ns=1.
likeli(As,ns)
```

```python
As_guess=1e-9
ns_guess=0.96
```

```python
old_As=As_guess
old_ns=ns_guess
for n in range(1000):
    old_loglikeli = loglikeli(old_As,old_ns)
    new_As = np.random.normal(old_As,1e-10)
    new_ns = np.random.normal(old_ns,0.01)
    new_loglikeli = loglikeli(new_As,new_ns)
    if (new_loglikeli > old_loglikeli):
        A.append([new_As,new_ns])
        old_loglikeli=new_loglikeli
        old_ns=new_ns
        old_As=new_As
    else:
        u = np.random.uniform(0.0,1.0)
        if (u < math.exp(new_loglikeli - old_loglikeli)):
            A.append([new_As,new_ns])
            old_loglikeli=new_loglikeli
            old_ns=new_ns
            old_As=new_As
        else:
            A.append([old_As,old_ns])
```

```python
print(A[-1])
```

```python
plt.hist(np.array(A)[:,1])
```

```python
As=A[-1][0]
ns=A[-1][1]
my_spectrum = get_spectrum(As,ns)
ell=np.zeros(len(my_spectrum))
for i in range(len(my_spectrum)):
    ell[i]=i
```
