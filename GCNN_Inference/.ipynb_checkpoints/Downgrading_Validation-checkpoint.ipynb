# Part 3. Kullback-Leibler Divergence

#### The Kullback-Leibler Divergence is a measure (though it does not satisfy the symmetry condition of metric axioms) that quantifies the difference between two probability distributions. It is defined as follows:
$$
D_\mathrm{KL}(P\parallel Q)=\int^\infty_{-\infty}\mathrm{d}xP(x)\mathrm{log}\left(\frac{P(x)}{Q(x)}\right)
$$
#### According to this definition, when $P(x)=Q(x)$, the divergence is equal to zero.

```python
from scipy.special import kl_div, rel_entr

hist_bins = 1000
hist_range = (-0.04, 0.06)
hist_T17_2048, _ = np.histogram(maps_T17_2048[0], bins=hist_bins, range=hist_range, density=True)
hist_T17_1024, _ = np.histogram(maps_T17_1024[0], bins=hist_bins, range=hist_range, density=True)
hist_T17_512, _ = np.histogram(maps_T17_512[0], bins=hist_bins, range=hist_range, density=True)
hist_T17_256, _ = np.histogram(maps_T17_256[0], bins=hist_bins, range=hist_range, density=True)
hist_T17_128, _ = np.histogram(maps_T17_128[0], bins=hist_bins, range=hist_range, density=True)

hist_Gauss_2048, _ = np.histogram(maps_Gauss_2048_mean, bins=hist_bins, range=hist_range, density=True)
hist_Gauss_1024, _ = np.histogram(maps_Gauss_1024_mean, bins=hist_bins, range=hist_range, density=True)
hist_Gauss_512, _ = np.histogram(maps_Gauss_512_mean, bins=hist_bins, range=hist_range, density=True)
hist_Gauss_256, _ = np.histogram(maps_Gauss_256_mean, bins=hist_bins, range=hist_range, density=True)
hist_Gauss_128, _ = np.histogram(maps_Gauss_128_mean, bins=hist_bins, range=hist_range, density=True)

hist_lognormal_2048, _ = np.histogram(maps_lognormal_2048_mean, bins=hist_bins, range=hist_range, density=True)
hist_lognormal_1024, _ = np.histogram(maps_lognormal_1024_mean, bins=hist_bins, range=hist_range, density=True)
hist_lognormal_512, _ = np.histogram(maps_lognormal_512_mean, bins=hist_bins, range=hist_range, density=True)
hist_lognormal_256, _ = np.histogram(maps_lognormal_256_mean, bins=hist_bins, range=hist_range, density=True)
hist_lognormal_128, _ = np.histogram(maps_lognormal_128_mean, bins=hist_bins, range=hist_range, density=True)

hist_T17_2048_mean, _ = np.histogram(map_T17_2048_mean, bins=hist_bins, range=hist_range, density=True)
hist_T17_1024_mean, _ = np.histogram(map_T17_1024_mean, bins=hist_bins, range=hist_range, density=True)
hist_T17_512_mean, _ = np.histogram(map_T17_512_mean, bins=hist_bins, range=hist_range, density=True)
hist_T17_256_mean, _ = np.histogram(map_T17_256_mean, bins=hist_bins, range=hist_range, density=True)
hist_T17_128_mean, _ = np.histogram(map_T17_128_mean, bins=hist_bins, range=hist_range, density=True)
## In order to avoid devision durch zero
epsilon = 1e-10
hist_T17_2048 += epsilon
hist_T17_1024 += epsilon
hist_T17_512 += epsilon
hist_T17_256 += epsilon
hist_T17_128 += epsilon

hist_Gauss_2048 += epsilon
hist_Gauss_1024 += epsilon
hist_Gauss_512 += epsilon
hist_Gauss_256 += epsilon
hist_Gauss_128 += epsilon

hist_lognormal_2048 += epsilon
hist_lognormal_1024 += epsilon
hist_lognormal_512 += epsilon
hist_lognormal_256 += epsilon
hist_lognormal_128 += epsilon

hist_T17_2048_mean += epsilon
hist_T17_1024_mean += epsilon
hist_T17_512_mean += epsilon
hist_T17_256_mean += epsilon
hist_T17_128_mean += epsilon

kl_div_2048_vs_1024 = np.sum(rel_entr(hist_T17_2048, hist_T17_1024))
kl_div_2048_vs_512 = np.sum(rel_entr(hist_T17_2048, hist_T17_512))
kl_div_2048_vs_256 = np.sum(rel_entr(hist_T17_2048, hist_T17_256))
kl_div_2048_vs_128 = np.sum(rel_entr(hist_T17_2048, hist_T17_128))

kl_div_2048_Gauss_vs_T17 = np.sum(rel_entr(hist_Gauss_2048, hist_T17_2048_mean))
kl_div_1024_Gauss_vs_T17 = np.sum(rel_entr(hist_Gauss_1024, hist_T17_1024_mean))
kl_div_512_Gauss_vs_T17  = np.sum(rel_entr(hist_Gauss_512, hist_T17_512_mean))
kl_div_256_Gauss_vs_T17  = np.sum(rel_entr(hist_Gauss_256, hist_T17_256_mean))
kl_div_128_Gauss_vs_T17  = np.sum(rel_entr(hist_Gauss_128, hist_T17_128_mean))

kl_div_2048_lognormal_vs_T17 = np.sum(rel_entr(hist_lognormal_2048, hist_T17_2048_mean))
kl_div_1024_lognormal_vs_T17 = np.sum(rel_entr(hist_lognormal_1024, hist_T17_1024_mean))
kl_div_512_lognormal_vs_T17  = np.sum(rel_entr(hist_lognormal_512, hist_T17_512_mean))
kl_div_256_lognormal_vs_T17  = np.sum(rel_entr(hist_lognormal_256, hist_T17_256_mean))
kl_div_128_lognormal_vs_T17  = np.sum(rel_entr(hist_lognormal_128, hist_T17_128_mean))

kl_div_2048_Gauss_vs_lognormal = np.sum(rel_entr(hist_Gauss_2048, hist_lognormal_2048))
kl_div_1024_Gauss_vs_lognoraml = np.sum(rel_entr(hist_Gauss_1024, hist_lognormal_1024))
kl_div_512_Gauss_vs_lognormal  = np.sum(rel_entr(hist_Gauss_512, hist_lognormal_512))
kl_div_256_Gauss_vs_lognormal  = np.sum(rel_entr(hist_Gauss_256, hist_lognormal_256))
kl_div_128_Gauss_vs_lognormal  = np.sum(rel_entr(hist_Gauss_128, hist_lognormal_128))

kl_div_2048_vs_1024_mean = np.sum(rel_entr(hist_T17_2048_mean, hist_T17_1024_mean))
kl_div_2048_vs_512_mean = np.sum(rel_entr(hist_T17_2048_mean, hist_T17_512_mean))
kl_div_2048_vs_256_mean = np.sum(rel_entr(hist_T17_2048_mean, hist_T17_256_mean))
kl_div_2048_vs_128_mean = np.sum(rel_entr(hist_T17_2048_mean, hist_T17_128_mean))
```
