## Part 3. Angular PowerSpectra
#### We will compare the angular power spectrum measured from the outputs of the simulation pipeline with that from the T17 simulation.
#### This comparison is crucial, as any discrepancies could indicate issues with my pipeline's results.
#### Figures 5.5 and 5.6 correspond to the content presented here.

#### Before making comparisons, I would like to introduce the following correction factors.
#### A comprehensive discussion of these factors can be found in Takahashi et al.(2017) (https://arxiv.org/abs/1706.01472) and Appendix B of Halder et al.(2021) (https://arxiv.org/abs/2102.10177) .

(i)**Finite simulation-box-size effect**: The precision of the power spectrum is influenced by the limited size of simulation volume. Spatial frequenceis that exceed the limit of Nyquist frequency, defined as $k<k_\mathrm{Nyq}=\frac{2\pi}{L}$, manifiest as aliased frequncies within the observable range due to the discretized nature of the numerical simulation grid. 

(ii)**Finite lens-shell effect**: When employing the ray tracing method via lens shells, it becomes imperative to account for the finite thickness inherent to each individual lens shells. This phenomena significantly influences the characteristics of the power spectrum. In this regard, Takahashi et al.(2017) offers a fitting formula obtained through the convolution of power spectrum and window function of the lens shells, yielding the subsequent expression.
$$
P(k,a)\to\frac{(1+c_1k^{-\alpha_1})^{\alpha_1}}{(1+c_2\alpha^{-\alpha_2})^{\alpha_3}}P_\mathrm{3D}(k,a)
$$
where $c_1=9.5171\times 10^{-4}$, $c_2=5.1543\times 10^{-3}$, $\alpha_1=1.3063$, $\alpha_2=1.1475$ and $\alpha_3=0.62793$.

(iii)**Finite angular resolution of sky-maps**: The measured angular power spectrum $C_l$ demonstrates an underestimation in relation to the corresponding theoretical spectrum. This discrepancy is primarily attributed to the inherent angular resolution limitations of the utilized Healpix map. In response to this, Takahashi et al.(2017) suggest the incorporation of a damping factor at smaller scales to effectively mitigate this angular resolution effect. This proposition can be concisely expressed as follows:
$$
C_l\to\frac{C_l}{1+\left(\frac{l}{l_\mathrm{max}}\right)^2}
$$
where $l_\mathrm{max}=1.6\cdot NSIDE$

```python
limber_with_shellthick = Limber_with_shellthick()
limber_with_FRE = Limber_with_FRE()
limber_with_both = Limber_with_both()
limber_without_both = Limber_without_both()
```

```python
Omega_m = 0.279
sigma_8 = 0.82
Cl_with_shellthick = limber_with_shellthick.Cl(Omega_m, sigma_8)[:,1]
Cl_with_FRE = limber_with_FRE.Cl(Omega_m, sigma_8)[:,1]
Cl_with_both = limber_with_both.Cl(Omega_m, sigma_8)[:,1]
Cl_without_both = limber_without_both.Cl(Omega_m, sigma_8)[:,1]
```

```python
ell_original  = np.linspace(1,len(Cl_original_mean),len(Cl_original_mean),dtype=np.int32)
ell_measured  = np.linspace(1,len(Cl_with_shellthick),len(Cl_with_shellthick),dtype=np.int32)
```

```python
plt.figure(figsize=(10,10))
plt.errorbar(ell_original,ell_original*(ell_original+1)*Cl_original_mean, Cl_original_var,label='T17 measured Power Spectrum')
plt.plot(ell_measured, ell_measured*(ell_measured+1)*Cl_without_both, label='Theoretical Power Spectrum without any Correction factors')
plt.plot(ell_measured, ell_measured*(ell_measured+1)*Cl_with_FRE, label='Theoretical Power Spectrum with Finite Angular Resolution')
plt.plot(ell_measured, ell_measured*(ell_measured+1)*Cl_with_shellthick, label='Theoretical Power Spectrum with Finite Lens-Shell Effect')
plt.plot(ell_measured, ell_measured*(ell_measured+1)*Cl_with_both, label='Theoretical Power Spectrum with both Correction factors')
#plt.xscale('log')
plt.yscale('log')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlim(0, 3000)
plt.ylim(0.00001,0.001)
plt.xlabel("$\ell$",fontsize=17)
plt.ylabel("$\ell(\ell+1)C_{\ell}$",fontsize=17)
plt.legend(fontsize=15)
plt.grid()
#plt.savefig('Plot/Angular_Power_Comparison.jpg',bbox_inches='tight')
plt.show();
```
